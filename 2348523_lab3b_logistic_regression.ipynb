```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
```

```python
data=pd.read_csv('/content/drive/MyDrive/Program_3b_dataset.csv')
data.head()
```

```python
data.isnull().sum()
```

```python
data.describe()
```

```python
data.dtypes
```

```python
df=pd.DataFrame(data)
df.head()
```

```python
df.shape
```

```python
X=df[['variance','skewness','curtosis','entropy']]
y=df['class']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=40)
```

```python
log_reg=LogisticRegression()
log_reg.fit(X_train,y_train)
```

```python
y_pred=log_reg.predict(X_test)
y_pred
```

```python
y_train_pred=log_reg.predict(X_train)
y_train_pred
```

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(y_test, y_pred)
print("Logistic Regression model accuracy for testing (in %):", acc*100)
```

```python
acc = accuracy_score(y_train, y_train_pred)
print("Logistic Regression model accuracy for training (in %):", acc*100)
```

```python
from sklearn.metrics import confusion_matrix ,classification_report
import seaborn as sn
cnf=confusion_matrix(y_test,y_pred)
cnf

```

```python
sn.heatmap(cnf, annot=True, fmt='d',linewidths=2)
plt.title('Confusion Matrix', fontsize=17)
plt.show()
```

```python
print("Classification report")
print(classification_report(y_test,y_pred))
```
