```python
num_df=data.drop(['Month', 'VisitorType', 'Weekend', 'Revenue'], axis=1)
num_df
```

```python
x=num_df
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
```

```python
#Standardise the data
sc=StandardScaler()
sc_data=sc.fit_transform(x)
sc_data
```

```python
# Determine the optimal number of clusters using the elbow method
k_value= range(1,11)
inertia = []
#features_clustering = ['Administrative', 'Administrative_Duration', 'Informational', 'Informational_Duration',   'ProductRelated', 'ProductRelated_Duration', 'BounceRates', 'ExitRates', 'PageValues', 'SpecialDay', 'OperatingSystems', 'Browser', 'Region', 'TrafficType']
for k in k_value:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(sc_data)
    inertia.append(kmeans.inertia_)
```

```python
# Plot the elbow method graph

plt.figure(figsize=(8,6))
plt.plot(k_value, inertia, marker='*')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
plt.show()
```

The elbow method is a graphical method for finding the optimal K value in a k-means clustering algorithm.
over here We selected 2 or 3 as potential optimal number of clusters because elbow plot is occurring around the number of clusters equal to 2 or 3 , it suggests that 2 or 3  clusters might be a suitable choice for K-means clustering.

```python
optimal_k =3

# Perform K-means clustering with optimal_k
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
data['Cluster'] = kmeans.fit_predict(sc_data)
```

```python
 #Replace 'Region' and 'Traffic' with your chosen features
import seaborn as sn
plt.figure(figsize=(10, 6))
sn.scatterplot(x='Region', y='TrafficType', hue='Cluster', data=data, palette='viridis', s=50)
plt.title('K-means Clustering')
plt.xlabel('Region')
plt.ylabel('TrafficType')
plt.show()
```

Hierarchial Clustering
It is a popular method in data analysis and machine learning for grouping similar data points into clusters based on their similarity or dissimilarity.

```python
# Import necessary library for hierarchical clustering
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt

# Select a subset of the data (first 100 data points)
subset_data = sc_data[:100, :]

# Perform hierarchical clustering on the subset
hierarchical_clusters = linkage(subset_data, method='ward')

# Plot the dendrogram for the subset
plt.figure(figsize=(12, 6))
dendrogram(hierarchical_clusters)
plt.title('Hierarchical Clustering Dendrogram (Subset of 100 Data Points)')
plt.xlabel('Data Points')
plt.ylabel('Distance')
plt.show()
```

```python
#sillhoutte coefficient for both clustering technique
silhouette_avg_kmeans = silhouette_score(sc_data, data['Cluster'])
print(f"Silhouette Score for K-means Clustering: {silhouette_avg_kmeans}")
```

```python
# Use AgglomerativeClustering for hierarchical clustering
from sklearn.cluster import AgglomerativeClustering

# Assume 'optimal_clusters_hierarchical' is the optimal number of clusters identified
optimal_clusters_hierarchical = 100

# Perform hierarchical clustering with optimal_clusters_hierarchical
hc = AgglomerativeClustering(n_clusters=optimal_clusters_hierarchical, linkage='ward')
data['Hierarchical_Cluster'] = hc.fit_predict(sc_data)

# Calculate silhouette score for Hierarchical clustering
silhouette_avg_hierarchical = silhouette_score(sc_data, data['Hierarchical_Cluster'])
print(f"Silhouette Score for Hierarchical Clustering: {silhouette_avg_hierarchical}")
```

in the above conclusion the values of

K-means Clustering: Silhouette Score is approximately 0.3164. This suggests that the clusters formed by K-means are relatively well-separated and distinct.

Hierarchical Clustering: Silhouette Score is approximately 0.1001. This indicates that the clusters formed by hierarchical clustering are less well-separated compared to K-means. It may suggest that the hierarchical clustering method has produced clusters with more overlap or less clear boundaries.
